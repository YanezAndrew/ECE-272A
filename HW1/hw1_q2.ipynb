```python
import openai
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

# Get API key from environment variable
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
def call_llm(prompt):
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "user", "content": "You are a python coding assistant. "},
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content
```

```python
# Read prompt from the text file
with open('hw1_q2_prompt.txt', 'r') as file:
    prompt = file.read()
```

```python
generated_code = call_llm(prompt)
print(generated_code)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

df = pd.read_csv('./autograder/source/housing.csv')
numeric_cols = df.select_dtypes(include=[np.number])
missing_col = numeric_cols.columns[numeric_cols.isnull().any()].tolist()[0]
median_value = numeric_cols[missing_col].median()
imputer = KNNImputer(n_neighbors=10)
imputed_values = imputer.fit_transform(numeric_cols)
imputed_value_row_291 = imputed_values[291, numeric_cols.columns.get_loc(missing_col)]
answer1 = [missing_col, median_value, imputed_value_row_291]
print(answer1)
```
